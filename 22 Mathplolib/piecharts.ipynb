```python
import matplotlib.pyplot as plt 
```

```python
slices=[60,40]
plt.title("Dwija Piechart")
lab=['dwija','krishu']
plt.pie(slices,labels=lab)
```

```python
x=[10,23,43,50]
labels=['dwija','krishu','mummma','pappa']
colors=['blue','black','pink','green']
plt.pie(x,labels=labels,colors=colors)
plt.show()
```

```python
explode=[0,0.5,0,0]
plt.pie(x,labels=labels,colors=colors,explode=explode,
        shadow=True,autopct='%1.1f%%')
plt.show()
```
