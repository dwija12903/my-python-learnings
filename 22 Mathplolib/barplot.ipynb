```python
# counetr module from collection counts all total element occurence
from collections import Counter
with open('data.csv') as f:
    readercsv=csv.DictReader(f)
    row=next(readercsv)
    c=Counter()
    for i in readercsv:
        c.update(i['LanguagesWorkedWith'].split(';'))
    print(c) #count of all the elements
    print(c.most_common(15)) #gives most comman 15 languages
```

```python
lang=[]
popu=[]
for item in c.most_common(15):
    lang.append(item[0])
    popu.append(item[1]) 
plt.barh(lang,popu,label='salary')
plt.xlabel('Language')
plt.ylabel('Popularity of Languages')
plt.title('LAnguagues and their poularity')
```

```python
import pandas as pd
data=pd.read_csv('data.csv')
id=data['LanguagesWorkedWith']
c=Counter()
for i in id:
    c.update(i.split(';'))
datareq=c.most_common(15)
x=[]
y=[]
for a in datareq:
    x.append(a[0])
    y.append(a[1])
plt.barh(x,y,label="data")
plt.xlabel('Popularity of Language')
plt.style.use('fivethirtyeight')
```
